```python
# Install datasets Library
!pip install datasets

# Import Libraries
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_country'].value_counts().head(20)
```

```python
uk_jobs = df[df['job_country'] == 'United Kingdom']
```

```python
uk_jobs.head()
```

```python
uk_jobs = uk_jobs[uk_jobs['salary_year_avg'].notna()]
uk_jobs
```

```python
uk_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median', ascending=False)
```

```python
date_counts = df.job_posted_date.value_counts()
date_counts = date_counts.sort_index()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
monthly_counts = df.job_posted_month.value_counts().sort_index()
```

```python
plt.plot(monthly_counts.index, monthly_counts)
```

```python
job_counts = df.job_title_short.value_counts().sort_values(ascending=True)
```

```python
plt.barh(job_counts.index, job_counts)
plt.title('Postings by title')
plt.ylabel('Job Title')
plt.xlabel('Count')
```

```python
job_counts.plot(kind='barh')
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
df.groupby('job_title_short')['salary_year_avg'].mean().sort_values()
```
